```python
#Count of missing data
for col in missing_info:
    num_missing_rows = data[data[col].isnull()==True].shape[0]
    print("Missing rows in {} column is {}".format(col, num_missing_rows))
```

```python
#Percentageof missing data

for col in missing_info:
    percentage_miss = data[data[col].isnull()==True].shape[0]/data.shape[0]
    print("% Missing rows in {} column is {}".format(col, percentage_miss))
```
